```python
import numpy as np

# Load the land mask
mask_path = "/Users/jmframe/qkrig/data/conus_land_mask.npy"
land_mask = np.load(mask_path)

# Print basic info
print("Shape:", land_mask.shape)
print("Data type:", land_mask.dtype)
print("Unique values:", np.unique(land_mask))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 3))
plt.imshow(land_mask, cmap='gray')
plt.title("CONUS Land Mask")
plt.colorbar(label="Mask Value")
plt.axis('off')  # Hide axis ticks
plt.tight_layout()
plt.show()
```
